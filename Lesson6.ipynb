Задача 1

    Даны значения величины заработной платы заемщиков банка (`salary`) и значения их поведенческого кредитного скоринга (`scoring`):

    salary = [35, 45, 190, 200, 40, 70, 54, 150, 120, 110],
    scoring = [401, 574, 874, 919, 459, 739, 653, 902, 746, 832],


    1. Найдите ковариацию этих двух величин: сначала без использования специальных функций, а затем с помощью функции `numpy.cov`.
     Полученные значения должны быть равны.

    2. Найдите коэффициент корреляции Пирсона: с помощью ковариации и средних квадратических отклонений двух признаков, 
    а затем с помощью специальных функций из библиотек `numpy` и `pandas`.
    




```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

X = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
Y = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])

#для начала посчитаем ковариацию простыми математическими операциями
MX = (35 + 45 + 190 + 200 + 40 + 70 + 54 + 150 + 120 + 110) / 10
MY = (401 + 574 + 874 + 919 + 459 + 739 + 653 + 902 + 746 + 832) / 10

#Z = X*Y
Z = 35*401+45*574+190*874+200*919+40*459+70*739+54*653+150*902+120*746+110*832

cov = Z/10 - MX*MY

print('ковариация простыми мат. операциями',cov)
print(' ')

#посчитаем с помощью специальных библиотек

MX = X.mean()
MY = Y.mean()

cov_1 = ((X - MX) * (Y - MY)).mean()
print('ковариация из первой формулы',cov_1)
print(' ')

cov_2 = (X*Y).mean() - MX*MY
print('ковариация из второй формулы',cov_2)
print(' ')

#посчитаем с помощью numpy cov
C = np.cov(X, Y)
print('с помощью библиотеки numpy')
print(C)
print(' ')

#рассчитаем коэффициент Пирсона

cp1 = cov / (X.std() * Y.std())
print('коэфф-т Пирсона, первый способ =', cp1)
print(' ')

cp2 = np.corrcoef(X, Y)
print('коэфф-т Пирсона, второй способ =')
print(cp2)
print(' ')
```

Задача 2

    Измерены значения IQ выборки студентов, 
    обучающихся в местных технических вузах:

    131, 125, 115, 122, 131, 115, 107, 99, 125, 111

    Известно, что в генеральной совокупности IQ распределен нормально. 
    Найдите доверительный интервал для математического ожидания
    с надежностью 0.95.
 



```python
sample = np.array([131, 125, 115, 122, 131, 115, 107, 99, 125, 111])

n = sample.shape[0]
m = sample.mean()
s = sample.std(ddof=1)

print ('n=',n, 'm=', m, 's=', s)

# alfa=1-0.95=0.05
# n-1=10-1=9
# t(0.05; 9)=2.2622 находим по таблице квантилей
t = 2.2622

lp = m - t * s / np.sqrt(n) #левый предел
pp = m + t * s / np.sqrt(n) #правый предел

print('доверительный интервал равен:')
print(lp, pp)
```

    Известно, что рост футболистов в сборной распределен нормально 
    с известной дисперсией 25. 
    На выборке объёма 27 выборочное среднее составило 174.2. 
    Найдите доверительный интервал для математического ожидания 
    с надежностью 0.95.



```python
Dx = 25
n = 27
Xv = 174.2
p = 0.92

# выборочный интервал равен (Хв-дельта; Хв+дельта)
# дельта=Ty*дельта/(корень из n)
# Ty находим из таблицы Фурье 2Ф(Ту)=Р/2
# 2Ф(Ту)=0.475 значит Ту=1.96
Ty = 1.96
Delta = (Ty * np.sqrt(Dx))/ np.sqrt(n)
print('доверительный интервал равен')
print(Xv - Delta, Xv + Delta)
```
